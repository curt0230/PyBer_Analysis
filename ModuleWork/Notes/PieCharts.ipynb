```python
# Import dependencies and check matplotlib version
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Pie chart using MATLAB
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
colors = ["slateblue","magenta","lightblue","green","yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

#length of tuple = wedges in pie, value to explode is fraction of radius
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)

plt.subplots(figsize=(8,8))

plt.pie(y_axis, 
    explode=explode_values,
    colors=colors,
    labels=x_axis, 
    autopct='%.1f%%')

plt.show()
```

```python
#Pie chart with OO interface
explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)

theme = ["slateblue","magenta","lightblue","green","yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

fig, ax = plt.subplots(figsize=(8,8))

ax.pie(y_axis,
    labels=x_axis,
    explode=explode,
    shadow=True,
    startangle=90,
    colors=theme,
    autopct="%.1f")

ax.invert_xaxis()

plt.show()
```
