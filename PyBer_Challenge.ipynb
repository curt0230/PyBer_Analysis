### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_count
```

```python
# 2. Get the total drivers for each city type
total_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
total_driver_count
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"] 
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fares / total_ride_count
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fares / total_driver_count
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
data = {"Total Rides":total_ride_count,
       "Total Drivers":total_driver_count,
       "Total Fares":total_fares,
       "Average Fare per Ride":avg_fare_per_ride,
       "Average Fare per Driver":avg_fare_per_driver}

pyber_summary_df = pd.concat(data,axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
import dataframe_image as dfi #for printing data frame to image

#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

dfi.export(pyber_summary_df, 'analysis/pyber_summary.png')
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Read the merged DataFrame
pyber_data_df
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# Note: double brackets on "fare" creates object as data frame
fares_by_type_date_df = pyber_data_df.groupby(["date","type"]).sum()[["fare"]]
fares_by_type_date_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_by_type_date_df = fares_by_type_date_df.reset_index()
fares_by_type_date_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

fares_pivot = fares_by_type_date_df.pivot(index="date",columns="type",values="fare")
fares_pivot
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_pivot.index = pd.to_datetime(fares_pivot.index)
fares_pivot.info()
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fares_pivot_filtered = fares_pivot.loc['2019-01-01':'2019-04-29']
fares_pivot_filtered
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fares_pivot_filtered.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_pivot_week = fares_pivot_filtered.resample("W").sum()
fares_pivot_week = fares_pivot_week.loc['2019-01-01':'2019-04-29']
fares_pivot_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import numpy as np

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# plt.rcParams['font.size'] = '10'

x_axis = fares_pivot_week.index
tick_locations = [value for value in x_axis.month] #create a tick location using a list comprehension

rural = fares_pivot_week.Rural
suburban = fares_pivot_week.Suburban
urban = fares_pivot_week.Urban

plt.subplots(figsize=(20,6))

#plot months and fares
plt.plot(x_axis, rural, label="Rural")
plt.plot(x_axis, suburban, label="Suburban")
plt.plot(x_axis, urban, label="Urban")

# plt.plot(x_axis, rural, label="Rural", linewidth=2)
# plt.plot(x_axis, suburban, label="Suburban", linewidth=2)
# plt.plot(x_axis, urban, label="Urban", linewidth=2)

plt.xlabel("2019")
plt.ylabel("Fare ($USD)")
plt.xticks(["2019-01-01","2019-02-01","2019-03-01","2019-04-01"], ["Jan","Feb","Mar","Apr"]) #pass the tick_locations array and the labels from the df
plt.title("Total Fare by City Type")
plt.grid(color="lightgrey")

plt.legend(loc="center")

plt.savefig("analysis/total_fare_by_city_type")
plt.show()
```
